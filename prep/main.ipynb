```python
import pandas as pd
import numpy as np
import requests
import json 
import random

poke_ids = pd.read_csv('pokeids.csv')
# np.array(poke_ids)[0][0]
```

```python
temp_arr = np.array(poke_ids)
# temp_arr = np.random.shuffle(temp_arr)

temp_list = []

for value in temp_arr:
    temp_list.append(value[0])

temp_list
```

```python
random.shuffle(temp_list)
```

```python
final_data = []
```

```python
for pid in temp_list:
    print('Curr pid:', pid)

    res = requests.get(f'https://pokeapi.co/api/v2/pokemon/{pid}/')
    response = json.loads(res.text)

    gen = 0

    if pid <= 151:
        gen = 1
    elif pid <= 251:
        gen = 2
    elif pid <= 386:
        gen = 3
    elif pid <= 493:
        gen = 4
    elif pid <= 649:
        gen = 5
    elif pid <= 721:
        gen = 6
    elif pid <= 809:
        gen = 7
    elif pid <= 905:
        gen = 8
    else:
        gen = 9

    height = response['height']
    weight = response['weight']
    name = response['forms'][0]['name']
    types = []

    for slot in response['types']:
        types.append(slot['type']['name'])

    res = requests.get(f'https://pokeapi.co/api/v2/pokemon-species/{pid}/')
    response = json.loads(res.text)

    for item in response['genera']:
        if item['language']['name'] == 'en':
            species = item['genus'].replace('Pokémon', ' ').strip()

    if len(name) > 10 or name.find('-') != -1:
        print(name)

    temp_data = {'id': pid, 'name': name, 'gen': gen, 'height': height, 'weight': weight, 'species': species, 'types': types}
    final_data.append(temp_data)
```

```python
final_data
```

```python
final_data_2 = {}

for pokemon in final_data:
    final_data_2[int(pokemon['id'])] = {
        'name': pokemon['name'],
        'gen': pokemon['gen'],
        'height': pokemon['height'],
        'weight': pokemon['weight'],
        'species': pokemon['species'],
        'types': pokemon['types']
    }

final_data_2
```

```python
class NpEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        if isinstance(obj, np.floating):
            return float(obj)
        if isinstance(obj, np.ndarray):
            return obj.tolist()
        return super(NpEncoder, self).default(obj)
```

```python
with open('poke_list.json', 'w') as outfile:
    json.dump(final_data, outfile, cls=NpEncoder)
```
